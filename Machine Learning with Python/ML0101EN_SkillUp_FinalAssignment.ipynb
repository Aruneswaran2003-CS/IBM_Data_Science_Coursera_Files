# Rain Prediction In Australia

## Importing Packages

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("./DataSets/Weather_Data.csv")
```

```python
df.head(5)
```

## Data Preprocessing

### One Hot Encoding

```python
from sklearn import preprocessing
df_processed = pd.get_dummies(data=df, columns=['RainToday','WindGustDir','WindDir9am','WindDir3pm'])
```

```python
df_processed.head(5)
```

```python
df_processed.replace(['No','Yes'],[0,1],inplace=True)
```

```python
df_processed.head(5)
```

```python
# drop the colummn date
df_processed.drop('Date',axis=1,inplace=True)
```

```python
df_processed
```

```python
df_processed=df_processed.astype(float)
```

## Feature Selection

```python
X=df_processed.drop(columns='RainTomorrow',axis=1)
Y=df_processed['RainTomorrow']
```

### Splitting the dataset into training and testing data for regression

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

### Building and training a model using Linear Regression 

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
LinearReg=lm.fit(x_train,y_train)
```

```python
#coefficent
print("Coefficent:",lm.coef_)
```

```python
#Intercept
print("Intercept:",lm.intercept_)
```

```python
yhat=lm.predict(x_test)
```

```python
yhat
```

### Calculating evaluation metrics for Linear Regression

```python
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score
MAE=mae(y_test,yhat)
MSE=mse(y_test,yhat)
R2=r2_score(y_test,yhat)
dictionary={"Metrics":["MAE","MSE","R2"],"Result":[MAE,MSE,R2]}
```

### Creating a final regression report/table of evaluation metrics

```python
from tabulate import tabulate
Report = pd.DataFrame(dictionary)
print(tabulate(Report, headers='keys',tablefmt='psql'))
```

### Building and training a model using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
neigh = KNeighborsClassifier(n_neighbors = 4).fit(x_train,y_train)
```

```python
yhat=neigh.predict(x_test)
```

```python
yhat
```

### Calculating evaluation metrics  for KNN

```python
from sklearn.metrics import jaccard_score as js
from sklearn.metrics import f1_score as f1
from sklearn.metrics import log_loss as ll
from sklearn.metrics import accuracy_score as As
```

```python
KNN_JACCARD_SCORE=js(y_test,yhat)
KNN_F1_SCORE=f1(y_test,yhat)
KNN_LOG_LOSS=ll(y_test,yhat)
KNN_ACCURACY_SCORE=As(y_test,yhat)
```

```python
print("KNN_JACCARD_SCORE:",KNN_JACCARD_SCORE)
print("KNN_F1_SCORE:",KNN_F1_SCORE)
print("KNN_LOG_LOSS:",KNN_LOG_LOSS)
print("KNN_ACCURACY_SCORE:",KNN_ACCURACY_SCORE)
```

###  Building and training a model using Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(criterion='entropy',max_depth=4)
tree.fit(x_train,y_train)
```

```python
yhat=tree.predict(x_test)
```

### Calculating evaluation metric for Decision Trees

```python
TREE_JACCARD_SCORE=js(y_test,yhat)
TREE_F1_SCORE=f1(y_test,yhat)
TREE_LOG_LOSS=ll(y_test,yhat)
TREE_ACCURACY_SCORE=As(y_test,yhat)
```

```python
print("TREE_JACCARD_SCORE:",TREE_JACCARD_SCORE)
print("TREE_F1_SCORE:",TREE_F1_SCORE)
print("TREE_LOG_LOSS:",TREE_LOG_LOSS)
print("TREE_ACCURACY_SCORE:",TREE_ACCURACY_SCORE)
```

### Building and training a model using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
LR=LogisticRegression(C=0.01,solver='liblinear').fit(x_train,y_train)
```

```python
yhat=LR.predict(x_test)
```

### Calculating evaluation metrics for Logistic Regression

```python
LR_JACCARD_SCORE=js(y_test,yhat)
LR_F1_SCORE=f1(y_test,yhat)
LR_LOG_LOSS=ll(y_test,yhat)
LR_ACCURACY_SCORE=As(y_test,yhat)
```

```python
print("LR_JACCARD_SCORE:",LR_JACCARD_SCORE)
print("LR_F1_SCORE:",LR_F1_SCORE)
print("LR_LOG_LOSS:",LR_LOG_LOSS)
print("LR_ACCURACY_SCORE:",LR_ACCURACY_SCORE)
```

### Building and training a model using SVM 

```python
from sklearn import svm
```

```python
SVM=svm.SVC(kernel='linear')
```

```python
SVM.fit(x_train,y_train)
```

```python
yhat=SVM.predict(x_test)
```

### Calculating evaluation metrics for SVM

```python
SVM_JACCARD_SCORE=js(y_test,yhat)
SVM_F1_SCORE=f1(y_test,yhat)
SVM_LOG_LOSS=ll(y_test,yhat)
SVM_ACCURACY_SCORE=As(y_test,yhat)
```

```python
print("SVM_JACCARD_SCORE:",SVM_JACCARD_SCORE)
print("SVM_F1_SCORE:",SVM_F1_SCORE)
print("SVM_LOG_LOSS:",SVM_LOG_LOSS)
print("SVM_ACCURACY_SCORE:",SVM_ACCURACY_SCORE)
```

### Creating a final classification report/table of evaluation metrics

```python
dictionary={"KNN":[KNN_JACCARD_SCORE,KNN_F1_SCORE,KNN_LOG_LOSS,KNN_ACCURACY_SCORE],
            "TREE":[TREE_JACCARD_SCORE,TREE_F1_SCORE,TREE_LOG_LOSS,TREE_ACCURACY_SCORE],
            "LR":[LR_JACCARD_SCORE,LR_F1_SCORE,LR_LOG_LOSS,LR_ACCURACY_SCORE],
            "SVM":[SVM_JACCARD_SCORE,SVM_F1_SCORE,SVM_LOG_LOSS,SVM_ACCURACY_SCORE]}
Report= pd.DataFrame(data=dictionary,index=['JACCARD_SCORE','F1_SCORE','LOG_LOSS','ACCURACY_SCORE'])
print(tabulate(Report,headers='keys',tablefmt='psql'))
```
